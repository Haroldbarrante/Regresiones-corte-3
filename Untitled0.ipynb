```python
import pandas as pd
datos=pd.read_excel("/content/drive/MyDrive/estadistica corte 2/datos_industrias.xlsx")
datos.head()
```

```python
# prompt: necesito que me realices una regrecion linela simple con "Tasa_Crecimiento" como x  y Y que sea "	Aumento_salario	" graficalo en dispercion

import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Suponiendo que 'datos' ya está definido como en el código anterior

# Preparar los datos
X = datos[['Tasa_Crecimiento']].values  # Asegúrate de que X sea una matriz NumPy 2D
y = datos['Aumento_salario'].values

# Crear y entrenar el modelo de regresión lineal
model = LinearRegression()
model.fit(X, y)

# Obtener los coeficientes
intercept = model.intercept_
slope = model.coef_[0]

# Realizar predicciones
y_pred = model.predict(X)


# Gráfico de dispersión con la línea de regresión
plt.scatter(X, y, label='Datos')
plt.plot(X, y_pred, color='red', label='Regresión Lineal')
plt.xlabel('Tasa_Crecimiento')
plt.ylabel('Aumento_salario')
plt.title('Regresión Lineal Simple')
plt.legend()
plt.show()

# Imprimir la ecuación de la recta
print(f"Ecuación de la recta: y = {slope:.2f}x + {intercept:.2f}")

# Calcular R^2
r_squared = model.score(X, y)
print(f"R^2: {r_squared:.2f}")
```

```python
# prompt: Generame una prueba de diferencia que tenga normalidad y en base a esta una prueba de diferencia t student,  analiza las variables que te permitan hacer las pruebas que te pedi y hazlo por favor, tambien aplica una prueba de Mann-Whitney U

import numpy as np
from scipy import stats

# Seleccionar las variables para las pruebas de diferencia.
# Basado en el análisis inicial, las únicas variables numéricas disponibles son 'Tasa_Crecimiento' y 'Aumento_salario'.
# Para realizar pruebas de diferencia entre grupos, necesitamos una variable categórica para definir los grupos.
# El archivo proporcionado no incluye una variable categórica explícita.
#
# **Asumiendo que deseamos comparar 'Aumento_salario' entre dos grupos hipotéticos**,
# necesitamos crear una variable categórica.
# Por ejemplo, podemos dividir las industrias en dos grupos basados en la 'Tasa_Crecimiento'
# (industrias de bajo crecimiento vs. alto crecimiento).
# Esto es una asunción y dependerá de si esta división es significativa para el análisis.

# Creamos una variable categórica 'Grupo_Crecimiento' basada en la mediana de 'Tasa_Crecimiento'
mediana_tasa = datos['Tasa_Crecimiento'].median()
datos['Grupo_Crecimiento'] = np.where(datos['Tasa_Crecimiento'] <= mediana_tasa, 'Bajo Crecimiento', 'Alto Crecimiento')

# Ahora podemos realizar pruebas de diferencia sobre 'Aumento_salario' entre 'Bajo Crecimiento' y 'Alto Crecimiento'.

# 1. Prueba de Normalidad para 'Aumento_salario' en cada grupo

grupo_bajo = datos[datos['Grupo_Crecimiento'] == 'Bajo Crecimiento']['Aumento_salario']
grupo_alto = datos[datos['Grupo_Crecimiento'] == 'Alto Crecimiento']['Aumento_salario']

print("--- Prueba de Normalidad (Shapiro-Wilk) ---")

# Prueba de Normalidad para el grupo de Bajo Crecimiento
stat_shapiro_bajo, p_shapiro_bajo = stats.shapiro(grupo_bajo)
print(f"Grupo 'Bajo Crecimiento':")
print(f"  Estadístico Shapiro-Wilk: {stat_shapiro_bajo:.4f}")
print(f"  Valor p: {p_shapiro_bajo:.4f}")
if p_shapiro_bajo > 0.05:
    print("  Los datos parecen seguir una distribución normal (no se rechaza H0).")
else:
    print("  Los datos no parecen seguir una distribución normal (se rechaza H0).")

# Prueba de Normalidad para el grupo de Alto Crecimiento
stat_shapiro_alto, p_shapiro_alto = stats.shapiro(grupo_alto)
print(f"\nGrupo 'Alto Crecimiento':")
print(f"  Estadístico Shapiro-Wilk: {stat_shapiro_alto:.4f}")
print(f"  Valor p: {p_shapiro_alto:.4f}")
if p_shapiro_alto > 0.05:
    print("  Los datos parecen seguir una distribución normal (no se rechaza H0).")
else:
    print("  Los datos no parecen seguir una distribución normal (se rechaza H0).")

print("-" * 35)

# 2. Prueba de Diferencia (Prueba t de Student o Mann-Whitney U)
# La elección entre la prueba t de Student y la prueba U de Mann-Whitney depende de los resultados
# de la prueba de normalidad y la homogeneidad de varianzas (aunque esta última es menos crítica para la t).
# Si ambos grupos son normales y tienen varianzas similares, se usa la t de Student.
# Si la normalidad no se cumple en uno o ambos grupos, o las varianzas son muy diferentes,
# se prefiere la prueba U de Mann-Whitney (prueba no paramétrica).

# Realizamos la prueba U de Mann-Whitney ya que es robusta ante la falta de normalidad.
# También realizaremos la prueba t de Student condicionalmente o como ejemplo.

print("--- Pruebas de Diferencia ---")

# Prueba U de Mann-Whitney
# Esta prueba no asume normalidad.
stat_mannwhitney, p_mannwhitney = stats.mannwhitneyu(grupo_bajo, grupo_alto)

print("\nPrueba U de Mann-Whitney:")
print(f"  Estadístico U: {stat_mannwhitney:.4f}")
print(f"  Valor p: {p_mannwhitney:.4f}")

if p_mannwhitney < 0.05:
    print("  Hay una diferencia estadísticamente significativa en el 'Aumento_salario' entre los grupos (se rechaza H0).")
else:
    print("  No hay una diferencia estadísticamente significativa en el 'Aumento_salario' entre los grupos (no se rechaza H0).")


# Prueba t de Student (asume normalidad y, por defecto en scipy, no asume igualdad de varianzas - Welch's t-test)
# Si los resultados de Shapiro-Wilk sugieren normalidad en ambos grupos (p > 0.05),
# se podría considerar la prueba t de Student. Aquí la aplicamos como ejemplo.

# Prueba de Levene para homogeneidad de varianzas (opcional, pero útil para la t de Student clásica)
# stat_levene, p_levene = stats.levene(grupo_bajo, grupo_alto)
# print(f"\nPrueba de Levene (Homogeneidad de Varianzas): Estadístico={stat_levene:.4f}, p={p_levene:.4f}")
# if p_levene < 0.05:
#     print("  Las varianzas son significativamente diferentes (se rechaza H0 de igualdad de varianzas).")
# else:
#     print("  Las varianzas son similares (no se rechaza H0).")


# Realizar Prueba t de Student (por defecto usa Welch's t-test, que no asume igualdad de varianzas)
# Si p_levene > 0.05, se podría usar equal_var=True para la t de Student clásica.
stat_ttest, p_ttest = stats.ttest_ind(grupo_bajo, grupo_alto, equal_var=False) # equal_var=False para Welch's t-test

print("\nPrueba t de Student (Welch's T-test):")
print(f"  Estadístico t: {stat_ttest:.4f}")
print(f"  Valor p: {p_ttest:.4f}")

if p_ttest < 0.05:
    print("  Hay una diferencia estadísticamente significativa en el 'Aumento_salario' entre los grupos (se rechaza H0).")
else:
    print("  No hay una diferencia estadísticamente significativa en el 'Aumento_salario' entre los grupos (no se rechaza H0).")

print("-" * 35)

# Análisis y Conclusiones:
# 1. **Normalidad:** Se verificó la normalidad de 'Aumento_salario' en cada grupo ('Bajo Crecimiento' y 'Alto Crecimiento')
#    utilizando la prueba de Shapiro-Wilk. Los valores p de esta prueba indican si se puede asumir normalidad (p > 0.05)
#    o si se desvían significativamente de la normalidad (p < 0.05). Es crucial revisar estos p-valores para
#    validar el uso de pruebas paramétricas como la t de Student.
#
# 2. **Prueba U de Mann-Whitney:** Esta prueba no paramétrica compara las medianas (o más precisamente, las distribuciones)
#    de 'Aumento_salario' entre los dos grupos. Es la prueba recomendada si los datos no cumplen el supuesto de normalidad.
#    Un p-valor < 0.05 indica una diferencia estadísticamente significativa entre los grupos en cuanto al aumento de salario.
#
# 3. **Prueba t de Student (Welch's T-test):** Esta prueba paramétrica compara las medias de 'Aumento_salario' entre los dos grupos.
#    La versión de Welch es preferible si las varianzas de los grupos son diferentes (lo cual es común).
#    Solo es apropiada si el supuesto de normalidad se cumple (o si el tamaño de la muestra es grande,
#    por el Teorema del Límite Central). Un p-valor < 0.05 indica una diferencia estadísticamente significativa
#    entre las medias de aumento de salario de los grupos.
#
# La elección final de la prueba de diferencia a interpretar (t de Student o Mann-Whitney) debe basarse en los resultados
# de la prueba de normalidad. Si al menos un grupo no es normal (p_shapiro < 0.05), se debe confiar en la prueba U de Mann-Whitney.
# Si ambos grupos son normales (p_shapiro > 0.05), la prueba t de Student es apropiada y puede tener más potencia.
# La interpretación de si existe una diferencia significativa se basa en comparar el valor p de la prueba elegida
# con el nivel de significancia (comúnmente 0.05).
```
